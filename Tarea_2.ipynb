# Problema A, modelo AR

```r
zt=rnorm (100, 0, 1)
m=1 #Numero de parametros para phi
n=1 #Numero de parametros para theta
aux<-(m+1):100
X1=numeric(100)
```

```r
# AR(1).1: X(t)=Z(t)+0.8X(t−1)
phis=c(1, 0.8) #Parametros phis
for(i in aux) {
X1[i]=t(phis)%*%c(zt[i],X1[i-1])
}
AR1.1=ts(X1, start=1, frequency=1)
plot(AR1.1, main='AR(1).1: X(t)=Z(t)+0.8X(t−1)')
acf(AR1.1)
pacf(AR1.1)
```
